```python
import json
import glob
```

```python
###常识判断
def extract_common_sense_questions():
    # 定义输入路径和输出路径
    input_path = '/kaggle/input/dataset111/*.json'
    output_path = '/kaggle/working/Changshipanduan.json'
    
    # 初始化存储容器
    all_questions = []
    
    # 遍历所有JSON文件
    for file_path in glob.glob(input_path):
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
            # 筛选常识判断题
            filtered = [q for q in data.get("questions", []) if q.get("type") == "常识判断"]
            all_questions.extend(filtered)
    
    # 保存到可写目录
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump({"questions": all_questions}, f, ensure_ascii=False, indent=2)

if __name__ == "__main__":
    extract_common_sense_questions()
    print("提取完成！输出文件保存在 /kaggle/working 目录")
```
